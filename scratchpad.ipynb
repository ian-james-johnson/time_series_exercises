```python
import numpy as np
import pandas as pd
import os

# Used to obtain information from websites
import requests
```

```python
# This cell can be used as a blue-print for a function

# Create an empty list that will be appended at each iteration
items_list = []
# Define the url where the data is kept
base_url = "https://python.zgulde.net/api/v1/items"
# Define the response from the request
response = requests.get(base_url)
# Convert the response to JSON
data = response.json()
# get the max page so we can create a loop
n = data['payload']['max_page']
```

```python
data
```

```python
n
```

```python
# Create new to allow for calling page
base_url = 'https://python.zgulde.net/api/v1/items?page='

# Create a for loop from 1 to n
# Remember that the range is non-inclusive
for i in range(1, n+1):
    # The full url is the base url + string version of page number (for string addition)
    url = base_url + str(i)
    # Get the response
    response = requests.get(url)
    # Convert the response into a python dictionary
    data = response.json()
    # Get the content for this iteration
    page_items = data['payload']['items']
    # Add the content from this iteration to the previous iterations
    items_list += page_items
```

```python
# This pulls the first dictionary from the list of dictionaries
items_list[0]
```

```python
# Turn it into a dataframe
items = pd.DataFrame(items_list)
items.head()

# The previous steps can be repeated to get stores and sales
```
