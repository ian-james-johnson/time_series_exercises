```python
import numpy as np
import pandas as pd
import os

# Used to obtain information from websites
import requests
```

```python
# This cell can be used as a blue-print for a function

# Create an empty list that will be appended at each iteration
items_list = []
# Define the url where the data is kept
base_url = "https://python.zgulde.net/api/v1/items"
# Define the response from the request
response = requests.get(base_url)
# Convert the response to JSON
data = response.json()
# get the max page so we can create a loop
n = data['payload']['max_page']
```

```python
data
```

```python
n
```

```python
# Create new to allow for calling page
base_url = 'https://python.zgulde.net/api/v1/items?page='

# Create a for loop from 1 to n
# Remember that the range is non-inclusive
for i in range(1, n+1):
    # The full url is the base url + string version of page number (for string addition)
    url = base_url + str(i)
    # Get the response
    response = requests.get(url)
    # Convert the response into a python dictionary
    data = response.json()
    # Get the content for this iteration
    page_items = data['payload']['items']
    # Add the content from this iteration to the previous iterations
    items_list += page_items
```

```python
# This pulls the first dictionary from the list of dictionaries
items_list[0]
```

```python
# Turn it into a dataframe
items = pd.DataFrame(items_list)
items.head()

# The previous steps can be repeated to get stores and sales
```

```python
# Getting stors next, using the same procedure as before
base_url = 'https://python.zgulde.net'
# Add an endpoint
response = requests.get(base_url + '/api/v1/stores')
```

```python
# Now get the sales data
response = requests.get(base_url + '/api/v1/sales')
response.json()
```

```python
# Create variable for each iteration of an upcoming loop
end_point = response.json()['payload']['next_page']
end_point # We are on page 1, so the next page is 2
```

```python
# Find out how many pages there are
response.json()['payload']['max_page']
```

```python
# The output is missing page number, fill that in with max page using string concatenation
# The cell above gives max page, so concatenate these two cells' content
end_point[:-1]
```

```python
# This gives the last page
end_point_last_page = end_point[:-1] + str(response.json()['payload']['max_page'])
end_point_last_page
```

```python
# Lets look at the last page of the data
last_page = requests.get(base_url + end_point_last_page).json()['payload']['sales']
last_page
```

```python
# Cast as a boolean
bool(last_page)
```

```python
# Assign pages to the ititial sales page
pages = requests.get(base_url + '/api/v1/sales?page=1').json()
pages
```

```python
# Assign endpoint to the next page
endpoint = requests.get(base_url + '/api/v1/sales?page=1').json()['payload']['next_page']
endpoint
```

```python
# Lets use a while loop to make a simple version of the previous code
while endpoint:
    # Get the response
    response = requests.get(base_url + endpoint).json()['payload']
    # Change the endpoint throughout the loop
    endpoint = response['next_page']
    # Pass the content to pages
    pages += response['sales']
    
# Keep getting error in this cell
# TypeError: unsupported operand type(s) for +=: 'dict' and 'list'    
```

```python
# For loop version
pages = []
for i in range(1, 184):
    response = requests.get(base_url + '/api/v1/sales?page=' + str(i))
    sales = response.json()['payload']['sales']
    pages += sales
```

```python
# Turn the output of the for loop into a dataframe
sales = pd.DataFrame(pages)
sales.head()
```

```python
# Now we start combining everything
# Firts merge; sales and stores
sales_plus_stores = pd.merge(sales, stores, how='left', left_on='store', right_on='store_id')
sales_plus_stores.info()
```

```python
# Second merge
everything = pd.merge(sales_plus_stores, items, how='left', left_on='item', right_on='item_id')
everything.info()
```

```python
# save to csv
everything.to_csv('cached_everything.csv')
```
